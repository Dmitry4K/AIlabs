## Логическая регрессия

### Описание алгоритма

_Логистическая регрессия (Logistic regression)_ — метод построения линейного классификатора, позволяющий оценивать апостериорные вероятности принадлежности объектов классам.

В данной задаче существует всего два класса, которым может принадлежать тот или иной объект, поэтому для вычисления вероятности пренадлежности объекта к классу съедобыных грибов будем использовать логистическую функцию: 

**f(x) = 1/(1 + exp(-z))**, где **z = B0 + B1 * X1 + B2 * X2 + B3 * X3 + ... + Bn * Xn**, a **x** - объект, его свойства (значения атрибутов, сформированных в структуру данныx)
1. f(x) принимает значения от 0 до 1 и соответственно в данной задаче будет обозначать вероятность съедобности гриба 
2. Параметры Xi - значения каждого атрибута в подаваемой на обработку строке.
3. Параметры Bi - коэффициенты, которые получаются в ходе обучения модели.

Таким образом, чтобы отнести объект к какому классу достаточно посчитать значение f(x) и основываясь на данной вероятности выдавать ответ.

### Построение модели

Нахождение коэффициентов будем осуществлять методом градиентного спуска. Реализация приведена ниже

```python
class log_reg:
    def __init__(self, learning_rate=0.01, num_iter=5000, fit_intercept=True, verbose=False):
        self.lr = learning_rate
        self.num_iter = num_iter
        self.fit_intercept = fit_intercept
        self.verbose = verbose
    def __b_intercept(self, X):
        # добавляем 1 к вектору атрибутов
        intercept = np.ones((X.shape[0], 1))
        return np.concatenate((intercept, X), axis=1)
    
    def sigm(self,z):
        return 1/(1 + np.exp(-z))
    
    def loss(self, yp, y):
        return (-y* np.log(yp) - (1-y)*np.log(1-yp)).mean()
    
    def fit(self, X, Y):
        x = np.array(X)
        y = np.array(Y)
        if self.fit_intercept:
            x = self.__b_intercept(x)
        self.w = np.zeros(x.shape[1])
        
        for i in range(self.num_iter):
            z = np.dot(x, self.w)
            yp = self.sigm(z)
            gradient = np.dot(x.T, (yp-y)) / y.size
            self.w -= self.lr * gradient
            z = np.dot(x, self.w)
            yp = self.sigm(z)
            loss = self.loss(yp,y)
        return self
    
    def predict(self, X):
        x = np.array([np.array(X)])
        if self.fit_intercept:
            x = self.__b_intercept(x)
        z = np.dot(x, self.w)
        return self.sigm(z).round()
    
    def score(self, x_test, y_test):
        success = 0;
        for index in range(len(x_test)):
            value = x_test[index]
            answr = y_test[index]
            if answr == self.predict(value):
                success += 1
        return success/len(x_test)
```

## Дерево решений

### Описание алгоритма

_Дерево решений_ - это древовидная структура, похожая на блок-схему, где внутренний узел представляет функцию (или атрибут), ветвь представляет правило решения, а каждый конечный узел представляет результат Самый верхний узел в дереве решений называется корневым узлом.

В данном методе машинного обучения во время обучения модели строится дерево решений, по которому в последствии находится принадлежность объекта к определенному классу. Чтобы получить ответ достаточно спуститься вниз по дереву, соблюдая правила в каждой вершины. Спускаться нужно до того момента, пока в дереве закончатся правила.

### Построение модели

На каждом этапе построения дерева (построения вершины) нужно находить наиболее удачный признак. Признак считается наиболее удачным, если у него, среди других признаков наименьшая энтропия. Энтропию можно считать разными способами, однако в данной лабораторной работе будем использовать код из предыдущей работы и будем считать вероятности каждого значения признака в каждом атрибуте каждого класса и на основе этой информации будем выбирать наиболее удачный признак (тот, который обладает минимальным шансом встретится среди ядовитых грибов и максимальным встретится среди съедобных грибов)

Таким образом получаем алгоритм построения узла:

1. Получаем на вход некоторый набор данных, разделяем его на классы
2. По каждому признаку делим выборку и считаем в ней кол-во съедобных и несъедобных грибов
3. Находим признак, разделив данные по которому получим наименьшую энтропию

Дерево решений для данной задачи будет иметь следующий вид:

_[[index, value, answer], [edible_tree], [poisonous_tree]]_

```python
class tree_solution:
    def __init__(self, max_depth=10, train_x = None, train_y = None):
        self.depth = max_depth
        self.train_x = train_x
        self.train_y = train_y
        self.filters = []
    
    def fit(self, x, y):
        self.train_x = x
        self.train_y = y
        self.tree_root = self.build_vertex(self.depth, x, y)
        return self
    
    def score(self, x_test, y_test):
        success = 0;
        for index in range(len(x_test)):
            value = x_test[index]
            answr = y_test[index]
            if answr == self.predict(value):
                success += 1
        return success/len(x_test)
    
    def tree_step(self, tree, x):
        index = tree[0][0]
        value = tree[0][1]
        next_tree = tree[1] if x[index] == value else tree[2]
        return self.tree_step(next_tree, x) if len(next_tree) != 0 else 1 if next_tree == tree[1] else 0
       
    def predict(self, x):
        return self.tree_step(self.tree_root, x)
    
    def get_prob_map(self, train_x, train_y):
        result_map = {}
        classes_count = {}
        # посчитаем количество объектов всех возможных классов
        for i in range(len(train_y)):
            cur = train_y[i]
            if cur not in classes_count:
                classes_count[cur] = 1
            else:
                classes_count[cur] += 1
        # найдем вероятности всех атрибутов встретится в каждом классе
        for i in range(len(train_x)):
            row = train_x[i]
            ans = train_y[i]
            
            if ans not in result_map:
                result_map[ans] = []
                cur_list = result_map[ans]
                while len(cur_list) != len(row):
                    cur_list.append({})
            
            cur_list = result_map[ans]
            for j in range(len(row)):
                value = row[j]
                cur_dict = cur_list[j]
                if value not in cur_dict:
                    cur_dict[value] = 1
                else:
                    cur_dict[value] += 1
        for key in result_map:
            current_list = result_map[key]
            for i in range(len(current_list)):
                cur_dict = current_list[i]
                for value in cur_dict:
                    cur_dict[value] /= classes_count[key]
        
        return result_map
    
    def entrophy(self, p, q):
        eps = 0.0000001
        if p < eps and q > eps:
            return - q * np.log2(q)
        elif p > eps and q < eps:
            return -p * np.log2(p) 
        elif p > eps and q > eps:
            return - q * np.log2(q) -p * np.log2(p) 
        else:
            return 0
    
    def get_edible_count(self, y):
        count = 0
        for value in y:
            if value == 1:
                count += 1
        return count
    
    def get_filter(self, x, y):
        #получаем новый узел ?
        index = None
        key_filter = None
        min_value = 1000
        prob_map = self.get_prob_map(x,y)
        for ans in prob_map:
            ans_dict = prob_map[ans]
            for i in range(len(ans_dict)):
                cur_dict = ans_dict[i]
                for key in cur_dict:
                    if [i,key] not in self.filters:
                        _, accepted_y, _, rejected_y = self.separate_by_filter(i, key, x,y)
                        p,q = 0,0
                        la = len(accepted_y)
                        lb = len(rejected_y)
                        eca = self.get_edible_count(accepted_y)
                        ecb = self.get_edible_count(rejected_y)
                        entra = self.entrophy(eca/la, (la-eca)/la) if la != 0 else 0
                        entrb = self.entrophy(ecb/lb, (lb-ecb)/lb) if lb != 0 else 0
                        entr = la/(la+lb) * entra + lb/(la+lb) * entrb
                        if min_value > entr and entr != 0:
                            min_value = entr
                            index = i
                            key_filter = key
        #print(min_value)
        if index != None:
            self.filters.append([index, key_filter])
        return index, key_filter
    
    def separate_by_filter(self, index, value, x, y):
        accepted_x = []
        rejected_x = []
        accepted_y = []
        rejected_y = []
        for i in range(len(x)):
            cur_list = x[i];
            cur_answ = y[i]
            if cur_list[index] == value:
                accepted_x.append(cur_list)
                accepted_y.append(cur_answ)
            else:
                rejected_x.append(cur_list)
                rejected_y.append(cur_answ)
        #print(len(accepted_x))
        #print(len(rejected_x))
        #print()
        return accepted_x, accepted_y, rejected_x, rejected_y
    
    def build_vertex(self, depth,x,y):
        vertex = []
        edible_vertex, poison_vertex = [], []
        if(depth > 0):
            #получаем значение фильтра
            index, value = self.get_filter(x, y)
            if index != None:
                #строим поддеревья ?
                accepted_x, accepted_y, rejected_x, rejected_y = self.separate_by_filter(index, value,x,y)
                edible_vertex = self.build_vertex(depth-1, accepted_x, accepted_y)
                poison_vertex = self.build_vertex(depth-1, rejected_x, rejected_y)
                vertex = [[index, value], edible_vertex, poison_vertex] 
                           
        return vertex
```

### Random Forest

По сути, Random Forest является композицией (ансамблем) множества решающих деревьев, что позволяет снизить проблему переобучения и повысить точность в сравнении с одним деревом. Прогноз получается в результате агрегирования ответов множества деревьев. Тренировка деревьев происходит независимо друг от друга (на разных подмножествах), что не просто решает проблему построения одинаковых деревьев на одном и том же наборе данных, но и делает этот алгоритм весьма удобным для применения в системах распределённых вычислений.

```python
import numpy as np
from scipy.stats import mode
from sklearn import tree



class random_forest:

    def __init__(self, n_estimators=2, bootstrap=0.5):
        self.n_estimators = n_estimators
        self.bootstrap = bootstrap
        self.forest = []


    def fit(self, Xl, yl):
        X = np.array(Xl)
        y = np.array(yl)
        self.forest = []
        n_samples = len(y)
        n_sub_samples = round(n_samples*self.bootstrap)
        
        for i in range(self.n_estimators):
            X_subset = X[:n_sub_samples]
            y_subset = y[:n_sub_samples]

            tree_ = tree.DecisionTreeClassifier(max_depth=2)
            tree_.fit(X_subset, y_subset)
            self.forest.append(tree_)
        return self


    def predict(self, X):
        n_samples = X.shape[0]
        n_trees = len(self.forest)
        predictions = np.empty([n_trees, n_samples])
        for i in range(n_trees):
            predictions[i] = self.forest[i].predict(X)

        return mode(predictions)[0][0]


    def score(self, Xl, yl):
        X = np.array(Xl)
        y = np.array(yl)
        y_predict = self.predict(X)
        n_samples = len(y)
        correct = 0
        for i in range(n_samples):
            if y_predict[i] == y[i]:
                correct = correct + 1
        accuracy = correct/n_samples
        return accuracy
```
